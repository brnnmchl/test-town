```python
import random
import csv
from tkinter import *
```

```python
def clicked(filename,table):
    lbl.configure(text = "I just got clicked")

def create_effects(filename,table):
    with open('paramon.csv') as file:
        csv_reader = csv.reader(file, delimiter=',')
        for row in csv_reader:
            effect_table.append(row)
    return(effect_table)
    
def starting_effects(table,effects,used):
    for x in range(0,8):
        effects.append(table[x])
        used.append(table[x])
    del table[0:8]
    return(table,effects,used)

def new_round(table,effects,used):
    index1 = random.randrange(len(effects))
    active_effect = effects[index1]
    print("You use " + active_effect[1])
    effects[index1] = table[0]
    used.append(table[0])
    del table[0]
    return(table,effects,used)

def get_started():
    create_effects('paramon.csv',effect_table)
    lbl1.configure(text = str(len(effect_table)) + " effects set! ")

def new_button(name,btntxt,cmd,col,rw,btndel):
    name = Button(root, text = btntxt, command = cmd)
    name.grid(column=col,row=rw)
    btndel.destroy()
    return(name)

def new_lbl(lblname,lbltxt,col,rw):
    lblname = Label(root, text = lbltxt)
    lblname.grid(column=col,row=rw)
    return(lblname)
```

```python
effect_table = []
effect_set = []
used_effects = []
```

```python
root = Tk()
# root window title and dimension
root.title("What's a Paramon to do?")
# Set geometry (widthxheight)
root.geometry('350x200')

lbl1 = Label(root, text = "Let's get started!")
lbl1.grid(column=1, row=0)

btn1 = Button(root, text = "Set the effects table?", command=get_started)
btn1.grid(column=3, row=1)

btn2 = new_button("btn2","Shuffle the effects table?",lambda: [random.shuffle(effect_table), new_lbl("lbl2","Effects table shuffled!",1,2)],3,2,btn1)

#btn3 = new_button("btn3","Generate starting effects?",lambda: [starting_effects(effect_table,effect_set,used_effects),new_lbl("lbl3","Your starting effects are: ",1,3)],3,3,btn2)

root.mainloop()
```

```python
effect_table = create_effects('paramon.csv',effect_table)
print(len(effect_table))
```

```python
random.shuffle(effect_table)

starting_effects(effect_table,effect_set,used_effects)
print("Your starting effects are: ")
for stat in effect_set:
    print(stat[1])
print(len(effect_table) + len(used_effects))
```

```python
new_round(effect_table,effect_set,used_effects)
print("Your new set of effects are:")
for stat in effect_set:
    print(stat[1])
print ("There are " + str(len(effect_table)) + " effects remaining.")
```
