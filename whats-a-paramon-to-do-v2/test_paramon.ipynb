```python
import random
import csv
from tkinter import *
```

```python
def create_effects(filename,table):
    with open('paramon.csv') as file:
        csv_reader = csv.reader(file, delimiter=',')
        for row in csv_reader:
            effect_table.append(row)
    return(effect_table)
    
def starting_effects(table,effects,used):
    for x in range(0,8):
        effects.append(table[x])
        used.append(table[x])
    del table[0:8]
    return(table,effects,used)

def new_button(container,name,btntxt,cmd,col,rw,stick):
    name = Button(container, text = btntxt, command = cmd)
    name.grid(column=col,row=rw,sticky=stick)
    return(name)

def new_lbl(lblname,lbltxt,col,rw,colspan,stick,fnt_typ,fnt_sz):
    lblname.config(text = lbltxt,font=(fnt_typ,fnt_sz))
    lblname.grid(column=col,row=rw,columnspan=colspan,sticky=stick)
    return(lblname)

def new_round(table,effects,lbl,used,col,rw,colspan,stick):
    index1 = random.randrange(len(effects))
    active_effect = effects[index1]
    new_lbl(lbl,"You use " + active_effect[1],col,rw,colspan,stick,"OCR A Extended",12)
    effects[index1] = table[0]
    used.append(table[0])
    del table[0]
    return(table,effects,used)

def get_started():
    create_effects('paramon.csv',effect_table)
    result1 = Label(set_up, text = str(len(effect_table)) + " effects set! ",
                    font=("OCR A Extended",12))
    result1.grid(column=2,row=0,sticky=E)
    btn1.destroy()

def get_effects(effect_set):
    effect_list = ""
    for stat in effect_set:
        effect_list = effect_list + "\n" + str(stat[1])
    return(effect_list)

def make_csv(used,ent_name):
    filename = ent_name.get() + ".csv"
    with open(filename,"w") as csvfile:
        writer = csv.writer(csvfile)
        writer.writerows(used)
```

```python
effect_table = []
effect_set = []
used_effects = []
```

```python
root = Tk()
root.title("What's a Paramon to do?")
root.geometry('600x480')
#root.config(background="lightgoldenrod1")
title = Frame(root,bg="darkorange2",borderwidth=5,relief="ridge")
title.grid(column=0,row=0,pady=5,padx=5)
set_up = Frame(root,bg="darkorange2",borderwidth=5,relief="ridge")
set_up.grid(column=0,row=1,pady=5,padx=5,sticky=W)
fx_zone = Frame(root,bg="darkorange2",borderwidth=5,relief="ridge")
fx_zone.grid(column=0,row=2,pady=5,padx=5,sticky=W)
end_sesh = Frame(root,bg="darkorange2",borderwidth=5,relief="ridge")
end_sesh.grid(column=0,row=3,pady=5,padx=5,sticky=W)

greet = Label(title, text = "WHAT'S A PARAMON TO DO?!",
              font=("OCR A Extended",20),fg="navy",bg="lightgoldenrod1")
greet.grid(column=0,row=0)

lbl1 = Label(set_up, text = "Set the effects table?",font=("OCR A Extended",12))
lbl1.grid(column=0,row=0,sticky=W)
btn1 = Button(set_up, text = "Set", command=get_started)
btn1.grid(column=1,row=0)

lbl2 = Label(set_up, text = "Shuffle the effects table?",font=("OCR A Extended",12))
lbl2.grid(column=0,row=1,sticky=W)
result2 = Label(set_up, text = "")
btn2 = new_button(set_up,"btn2","Shuffle",
                  lambda:[random.shuffle(effect_table),
                    new_lbl(result2,"Effects table shuffled",2,1,1,E,"OCR A Extended",12),
                    btn2.destroy()],1,1,NW)

lbl3 = Label(fx_zone, text = "Generate starting effects?",font=("OCR A Extended",12))
lbl3.grid(column=0,row=0,sticky=W)
lbl5 = Label(fx_zone, text="",font=("OCR A Extended",12))
lbl6 = Label(fx_zone, text="",font=("OCR A Extended",12))
btn3 = new_button(fx_zone,"btn3","Generate",
                  lambda: [starting_effects(effect_table,effect_set,used_effects),
                    new_lbl(lbl5,"Your starting effects are: ",0,1,1,N,"OCR A Extended",12),
                    new_lbl(lbl6,get_effects(effect_set),0,2,1,N,"OCR A Extended",12),
                    btn3.destroy()],1,0,W)

lbl4 = Label(fx_zone, text = "Start first round?",font=("OCR A Extended",12))
lbl4.grid(column=0,row=1,sticky=NW)
effects_remain = Label(fx_zone, text = "")
used = Label(fx_zone, text = "")
btn4 = Button(fx_zone, text="Get effects",
              command=lambda:[new_round(effect_table,effect_set,used,used_effects,1,1,1,W),
                                lbl5.configure(text="Your new effects are: "),
                                lbl6.configure(text = get_effects(effect_set)),
                                new_lbl(effects_remain,"There are " + str(len(effect_table)) + " effects remaining.",
                                        2,1,1,NW,"OCR A Extended",12),
                                lbl4.configure(text="Next Round: ")])
btn4.grid(column=1,row=4,sticky=NW)

csv_lbl = Label(end_sesh, text="Enter filename: ",font=("OCR A Extended",12))
csv_lbl.grid(column=2,row=10,sticky=NW)
ent_filename = Entry()
ent_filename.grid(column=2,row=11,sticky=NW)
csv_created = Label(end_sesh,text="")
used_csv_btn = Button(end_sesh, text="Get used effects",
                      command=lambda: [make_csv(used_effects,ent_filename),
                                       new_lbl(csv_created,"CSV created in root folder!",
                                               2,11,1,NW,"OCR A Extended",12)])
used_csv_btn.grid(column=2,row=12,sticky=NW)

quit_btn = Button(end_sesh, text="All done",command=root.destroy)
quit_btn.grid(column=0,row=14,sticky=SW)

root.mainloop()
```
